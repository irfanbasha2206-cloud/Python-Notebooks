```python
# Framework

'''
A framework is a predefined structure or platform that helps developers build applications quickly by providing tools, templates, and reusable code.
Unlike a library, a framework calls your code (inversion of control).

Provides a skeleton for your application so you don’t have to handle everything manually.
Helps enforce best practices and organization in your project.

Key points:

Handles repetitive tasks like routing, database access, templates, or UI.
Can be web frameworks, GUI frameworks, or data/ML frameworks.

'''


# Popular Frameworks in Python

'''
1.  Web Frameworks
Django – Full-stack web framework, includes ORM, templates, authentication
Flask – Lightweight web framework, very flexible
FastAPI – Modern web framework for building APIs quickly
Bottle – Minimalist web framework


2.  GUI Frameworks
Tkinter – Built-in GUI toolkit for desktop apps
PyQt / PySide – Advanced GUI applications
Kivy – Cross-platform GUI for desktop and mobile


3. Data Science / ML Frameworks
Streamlit – Build interactive data apps quickly
Dash – Web apps for data visualization
TensorFlow – Deep learning framework
PyTorch – Machine learning and neural networks


4.  Game Development Frameworks
Pygame – 2D games development
Panda3D – 3D game engine

'''
```

```python
# Streamlit

'''
Streamlit is a Python framework used for building interactive web applications for data science and machine learning.
It allows you to create web apps quickly without needing HTML, CSS, or JavaScript.

It is lightweight and intuitive, mostly used for data visualization dashboards and prototype apps.

'''

# Uses of Streamlit

'''
Data Visualization Dashboards
-->  Easily display charts and graphs with Matplotlib, Seaborn, or Plotly.

Interactive Widgets
--> Use sliders, dropdowns, buttons to interact with data.

Machine Learning App Prototypes
--> Display predictions from ML models in real-time.

Rapid Prototyping
--> Quickly turn a Python script into a web app.

'''
```

```python
# Methods to use streamlit framework library

# 1. Install Streamlit
# pip install streamlit

# 2. Import Streamlit (alias) at end
# import streamlit as st 

# 3. Run the file in terminal
# streamlit run filename.py
```

```python
# Streamlit Functions Definitions

# 1. Text & Data Display
'''
st.write() → Displays anything (text, data, charts, etc.) automatically.
st.text() → Displays plain text (no formatting).
st.markdown() → Displays formatted text using Markdown.
st.latex() → Displays mathematical formulas in LaTeX format.
st.title() → Displays a large title.
st.header() → Displays a section header.
st.subheader() → Displays a smaller header.
st.caption() → Displays small descriptive text.
st.code() → Displays code with syntax highlighting.
st.json() → Displays JSON data in a structured format.
st.dataframe() → Shows interactive tables (sortable/filterable).
st.table() → Shows static tables (non-interactive).
st.metric() → Displays key values with indicators (like KPI).
'''

# 2. Charts & Visualization
'''
st.line_chart() → Displays a line chart.
st.bar_chart() → Displays a bar chart.
st.area_chart() → Displays an area chart.
st.scatter_chart() → Displays a scatter plot.
st.map() → Displays data points on a map.
st.pyplot() → Displays Matplotlib charts.
st.altair_chart() → Displays Altair charts.
st.plotly_chart() → Displays Plotly interactive charts.
st.vega_lite_chart() → Displays Vega-Lite charts.
'''

# 3. Input Widgets
'''
st.button() → Creates a clickable button.
st.checkbox() → Creates a checkbox input.
st.radio() → Creates radio buttons (single selection).
st.selectbox() → Dropdown for selecting one option.
st.multiselect() → Dropdown for selecting multiple options.
st.slider() → Slider to select numeric values.
st.select_slider() → Slider for selecting from options.
st.text_input() → Input field for single-line text.
st.number_input() → Input field for numbers.
st.text_area() → Input field for multi-line text.
st.date_input() → Input field for selecting a date.
st.time_input() → Input field for selecting time.
st.file_uploader() → Upload files.
st.color_picker() → Pick a color.
st.camera_input() → Capture image using camera.
'''

# 4. Media
'''
st.image() → Displays images.
st.audio() → Plays audio files.
st.video() → Plays video files.
📐 5. Layout & Structure
st.sidebar → Creates a sidebar for inputs/content.
st.columns() → Creates multiple columns layout.
st.container() → Groups elements together.
st.expander() → Expandable/collapsible section.
st.tabs() → Creates tabbed layout.
st.empty() → Placeholder to update content dynamically.
'''

# 6. Status & Feedback
'''
st.spinner() → Shows loading spinner.
st.progress() → Displays progress bar.
st.success() → Shows success message.
st.info() → Shows info message.
st.warning() → Shows warning message.
st.error() → Shows error message.
st.toast() → Shows temporary popup message.
st.balloons() → Displays celebration balloons 🎈
st.snow() → Shows snow animation 
'''

# 7. Advanced / App Control
'''
st.set_page_config() → Sets page title, icon, layout.
st.session_state → Stores data across interactions.
st.cache_data() → Caches data to improve performance.
st.cache_resource() → Caches heavy resources (models, DB).
st.rerun() → Reruns the app script.
st.stop() → Stops execution of the app.
st.experimental_get_query_params() → Gets URL parameters.
st.experimental_set_query_params() → Sets URL parameters.
'''
```
