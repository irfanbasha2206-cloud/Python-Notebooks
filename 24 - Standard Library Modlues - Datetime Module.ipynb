```python
#  datetime module 
'''
The datetime module is a standard library module in Python that provides classes and functions to work with dates and times.

It is not a built-in module (you need to import it).
It allows you to create, manipulate, format, and calculate with dates and times.

'''

# Main Classes in datetime
# The datetime module provides several important classes:

'''

| Class       | Description                                       |
| ----------- | ------------------------------------------------- |
| `date`      | Handles **year, month, day**                      |
| `time`      | Handles **hour, minute, second, microsecond**     |
| `datetime`  | Combines **date + time**                          |
| `timedelta` | Represents **difference** between two dates/times |
| `timezone`  | Handles **timezones**                             |
| `tzinfo`    | Base class for timezone information (advanced)    |

'''


import datetime as dt

# shows exact date time with microscopic seconds

now = dt.datetime.now()
print(now)                       
```

```python
# date Class Methods

'''
| Method                       | Description                                   |
| ---------------------------- | --------------------------------------------- |
| `today()`                    | Returns current local date                    |
| `fromtimestamp(ts)`          | Create date from timestamp                    |
| `fromisoformat(date_string)` | Parse ISO format string `YYYY-MM-DD`          |
| `replace(year, month, day)`  | Return new date with replaced values          |
| `weekday()`                  | Returns day of week (0=Monday … 6=Sunday)     |
| `isoweekday()`               | Returns ISO day of week (1=Monday … 7=Sunday) |
| `strftime(format)`           | Format date as string                         |


'''
# Create a date

import datetime as dt

d = dt.date(2025, 4, 5)
print(d)
```

```python
# Get today’s date

import datetime as dt

today = dt.date.today()
print(today) 
```

```python
# Access each attributes

import datetime as dt

print(today.year)
print(today.month)
print(today.day)
```

```python
# time Class Methods
'''
| Method                                       | Description                          |
| -------------------------------------------- | ------------------------------------ |
| `replace(hour, minute, second, microsecond)` | Return new time with replaced values |
| `strftime(format)`                           | Format time as string                |

'''

# Get today’s time

import datetime as dt

timenow = dt.datetime.now().time()
print(timenow)
```

```python
# Create time

from datetime import time

t = time(14, 30, 45)
print(t)
```

```python
# Access attributes

print(t.hour)
print(t.minute)
print(t.second)
```

```python
# datetime Class Methods

'''
| Method                                            | Description                                |
| ------------------------------------------------- | ------------------------------------------ |
| `now(tz=None)`                                    | Current local datetime (or timezone aware) |
| `utcnow()`                                        | Current UTC datetime                       |
| `today()`                                         | Current datetime (same as `now()`)         |
| `fromtimestamp(ts)`                               | Timestamp → datetime                       |
| `utcfromtimestamp(ts)`                            | Timestamp → UTC datetime                   |
| `fromisoformat(date_string)`                      | Parse ISO datetime string                  |
| `strptime(date_string, format)`                   | Parse string → datetime                    |
| `strftime(format)`                                | Format datetime → string                   |
| `replace(year, month, day, hour, minute, second)` | Modify datetime                            |
| `combine(date, time)`                             | Combine date + time into datetime          |
| `timestamp()`                                     | Datetime → POSIX timestamp                 |

'''
```

```python
# String Formatting time (strftime)

# Convert datetime → string

'''
Common    format codes
Code	   (Meaning)

%Y	      Year (2025) (Y - current year with century ) (y - current year without century) 
%m        Month (01–12)
%d	      Day
%H	      Hour (24h)
%M	      Minute
%S	      Second

'''

from datetime import datetime

now = datetime.now()

formatted = now.strftime("%d-%m-%Y %H:%M:%S")
print(formatted)
```

```python
# String Parsing time (strptime)

# Convert string → datetime

from datetime import datetime

date_str = "05-04-2025"
dt = datetime.strptime(date_str, "%d-%m-%Y")
print(dt)
```

```python
# Extracting days from the module
# " %A " -- to display the full day of the date
# " %a " -- to display the half day of the date
# display the day by current date 


import datetime as dt

day = dt.date.today()
print(day.strftime("%A"))
```

```python
# display the day by given date

import datetime as dt

day = dt.date(2001,6,22)
print(day.strftime("%A"))
```

```python
# using weekday() method 
# to find the position of day in a week

import datetime as dt

today = dt.date.today()
print(today.weekday())
```

```python
# timedelta (Date Arithmetic)

# Used to add/subtract only time

# timedelta Class Methods

'''
| Method / Operation                                                            | Description                           |
| ----------------------------------------------------------------------------- | ------------------------------------- |
| `timedelta(days, seconds, microseconds, milliseconds, minutes, hours, weeks)` | Create timedelta                      |
| `+ / -`                                                                       | Add or subtract timedelta to datetime |
| `total_seconds()`                                                             | Get total duration in seconds         |

'''

from datetime import datetime, timedelta

now = datetime.now()

future = now + timedelta(days=5)
past = now - timedelta(days=3)

print(future)
print(past)
```

```python
# if we want to add / sub month ar year
# use another module

import datetime as dt
from dateutil.relativedelta import relativedelta

today = dt.date.today()

future = today + relativedelta(months=2,years=4)
print(future)
```

```python
# Difference Between Dates

from datetime import date

d1 = date(2025, 4, 5)
d2 = date(2025, 4, 1)

diff = d1 - d2
print(diff)
```

```python
# Replace Values

from datetime import datetime

now = datetime.now()

new_dt = now.replace(year=2030)
print(new_dt)
```

```python
# timezone Class Methods

'''
| Method             | Description                    |
| ------------------ | ------------------------------ |
| `timezone(offset)` | Create timezone-aware datetime |
| `utcoffset()`      | Offset from UTC                |
| `tzname()`         | Name of the timezone           |

'''

# Timezone Handling (Basic)

from datetime import datetime, timezone

utc_now = datetime.now(timezone.utc)
print(utc_now)
```

```python
# Creating a timezone-aware datetime

from datetime import datetime, timezone, timedelta

# Create UTC+5:30 timezone
tz_ist = timezone(timedelta(hours=5, minutes=30))

# Current time in UTC+5:30
dt_ist = datetime.now(tz_ist)

print("Timezone-aware datetime:", dt_ist)
```

```python
# Real-World Example
# Age Calculator

from datetime import date

birth = date(2001, 6, 22)
today = date.today()

age = today.year - birth.year
print(age)
```
