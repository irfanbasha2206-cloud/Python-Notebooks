# Difference Between Compiler and Interpreter


'''
| Feature             | Compiler                                                             | Interpreter                                     |
| ------------------- | -------------------------------------------------------------------- | ----------------------------------------------- |
| **Definition**      | Translates the **whole program** into machine code before execution. | Translates **and executes code line by line**.  |
| **Execution**       | After compilation (produces executable file).                        | During interpretation (runs immediately).       |
| **Error Detection** | Shows **all errors after compilation**.                              | Shows **errors line by line** as code executes. |
| **Speed**           | Faster execution (once compiled).                                    | Slower execution (line-by-line).                |
| **Output**          | Produces an **independent executable file**.                         | No separate file; runs through the interpreter. |
| **Examples**        | C, C++, Java (compiled to bytecode then executed)                    | Python, Ruby, JavaScript                        |
| **Memory Use**      | Requires more memory for compiled code.                              | Requires less memory during execution.          |


```python
# Compiler

'''
A compiler is a program that translates the entire source code of a programming language into machine code (binary) before execution.

--> Converts high-level code → low-level machine code
--> Produces an executable file that the computer can run directly
--> Typical in languages like C, C++, and Java

'''

# Interpreter

'''
An interpreter is a program that reads a high-level programming language and executes it line by line.

--> It translates each line of code into machine code and immediately executes it.
--> Unlike a compiler, it does not produce a separate executable file.
--> Errors are detected at runtime, meaning as the program runs.

--> Python code (.py file) → compiled to bytecode (.pyc) → executed by Big ram memory commonly called as Python Virtual Machine (PVM) 
--> This is why Python programs run immediately and are easy to test line by line.
```

# Concepts of Python

```python

'''
1. Interpreted Language

Python code is executed line-by-line using an interpreter, unlike compiled languages (like C or C++).

This makes debugging easier and development faster.

2. High-Level Language

You don’t need to worry about low-level details like memory management.

3. Dynamically Typed

You don’t need to declare variable types.

4. Object-Oriented

Python supports classes and objects, enabling reusable and organized code.

5. Platform Independent

Code written once can run on Windows, Mac, Linux without changes.
```
