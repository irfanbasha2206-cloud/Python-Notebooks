```python
#factors ---> A factor of a number is a number that multiplies with another number to produce that number. 
n=int(input("Enter the number"))
for i in range(1,n+1):
    if n%i==0: 
        print(i)
```

```python
#count of factors
count=0
n=int(input("Enter the number"))
for i in range(1,n+1):
    if n%i==0:
        count=count+1
print(count)
```

```python
#factorial--> factorial is a function used to multiply a number by all the positive integers less than it.
#Example --> 6!-->6*5*4*3*2*1
fact=1
n=int(input("Enter the number"))
for i in range(1,n+1):
    fact=fact*i
print(fact)
    
```
