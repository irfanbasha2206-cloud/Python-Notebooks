```python
#bodmas Method 
a=int(input("Enter the 1 number: "))
b=int(input("Enter the 2 number: "))
c=int(input("Enter the 3 number: "))
d=(a+b)*c                              #first works bracket order values and then works multiplication  
print("the total answer is ",d)
```

```python
# alternate method
a=int(input("Enter the 1 number: "))
b=int(input("Enter the 2 number: "))
c=int(input("Enter the 3 number: "))
d=a+b
e=c*d                                    #assigning a new variable for next multiplication operator     
print("the total answer is ",e)
```
