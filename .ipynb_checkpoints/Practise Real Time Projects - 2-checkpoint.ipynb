```python
import pyttsx3
import datetime
import webbrowser

# Initialize engine
engine = pyttsx3.init()

def speak(text):
    print("Jarvis:", text)
    engine.say(text)
    engine.runAndWait()

def listen():
    # Replace microphone with text input (works in Jupyter)
    command = input("You: ")
    return command.lower()

def run_jarvis():
    speak("Hello boss, I am Jarvis.")

    while True:
        command = listen()

        if "time" in command:
            time = datetime.datetime.now().strftime("%H:%M")
            speak(f"The time is {time}")

        elif "open youtube" in command:
            webbrowser.open("https://youtube.com")
            speak("Opening YouTube")

        elif "open google" in command:
            webbrowser.open("https://google.com")
            speak("Opening Google")

        elif "date" in command:
            date = datetime.datetime.now().strftime("%Y-%m-%d")
            speak(f"Today's date is {date}")

        elif "exit" in command or "quit" in command:
            speak("Goodbye boss")
            break

        elif command.strip() == "":
            speak("I didn't catch that")

        else:
            speak("Sorry, I don't understand that yet")

# Run assistant
run_jarvis()
```

#  Cinematic Hacker Terminal (Matrix Style)

```python
import time
import random
import builtins  # important fix

def type_text(text):
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.03)
    print()

def choose_media():
    print("Choose The Media..!")

    platforms = [
        "Facebook",
        "Instagram",
        "YouTube",
        "WhatsApp",
        "Twitter (X)",
        "LinkedIn",
        "Snapchat",
        "TikTok"
    ]

    for i in platforms:
        print(i)
        time.sleep(1.2)

    hacker_input = builtins.input("Enter The Social Media: ")

    if hacker_input in platforms:
        print(hacker_input, "is loading.....!")
        time.sleep(1.0)
        return hacker_input
    else:
        print("Invalid choice!")
        return choose_media()

def hacking():
    type_text("Initializing system...")
    time.sleep(1)

    for i in range(5):
        fake = ''.join(random.choice("01ABCDEF") for _ in range(16))
        print("Decrypting:", fake)
        time.sleep(0.5)

    type_text("Access Granted ✔")
    type_text("Welcome, Agent Irfan 😎")

# Run program
hacking()
choose_media()
```

# AI Mood-Based Music Recommender

```python
import builtins

moods = {
    "happy": ["Shape of You", "Happy - Pharrell", "On Top of the World"],
    "sad": ["Let Her Go", "Someone Like You", "Fix You"],
    "motivated": ["Believer", "Lose Yourself", "Stronger"],
}

def show_moods():
    print("\nAvailable moods:")
    for mood in moods:
        print("-", mood)

def recommend():
    while True:
        show_moods()
        
        mood = builtins.input("\nEnter your mood: ").lower().strip()

        if mood in moods:
            print("\n🎧 Recommended Songs:")
            for song in moods[mood]:
                print("🎵", song)
            break  # stop after valid input
        else:
            print("\n❌ Invalid mood! Please choose from the list.")

# Run
recommend()
```
