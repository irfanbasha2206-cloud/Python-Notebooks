# Calculator Project

```python
def addition(a, b):
    return a + b

def subtraction(a, b):
    return a - b

def multiplication(a, b):
    return a * b

def division(a, b):
    return a / b

def floor_division(a, b):
    return a // b

def power(a, b):
    return a ** b


print("Welcome to my calculator")

while True:
    print("\nChoose your options")
    print("1 for add", "2 for sub", "3 for multiply", "4 for division",
          "5 for floor division", "6 for power", "7 for exit")

    useroption = input("Enter your option: ")

    if useroption == "7":
        print("Come back soon")
        break   # 🔴 stops the loop

    elif useroption in ["1", "2", "3", "4", "5", "6"]:
        try:
            a = int(input("Enter the first number: "))
            b = int(input("Enter the second number: "))

            if useroption == "1":
                print("Your added result is", addition(a, b))

            elif useroption == "2":
                print("Your subtracted result is", subtraction(a, b))

            elif useroption == "3":
                print("Your multiplied result is", multiplication(a, b))

            elif useroption == "4":
                if b == 0:
                    print("Cannot divide by zero!")
                else:
                    print("Your divided result is", division(a, b))

            elif useroption == "5":
                if b == 0:
                    print("Cannot divide by zero!")
                else:
                    print("Your floor division result is", floor_division(a, b))

            elif useroption == "6":
                print("Your power value result is", power(a, b))

        except ValueError:
            print("Please enter valid numbers!")

    else:
        print("Invalid option. Please try again.")
```

# Contact Book version - 2 📒

```python
def contact_book():
    contacts = {}

    # 🚨 Emergency Contacts
    search_contacts = {
        "police": "100",
        "fire": "101",
        "ambulance": "102",
        "women helpline": "1091",
        "child helpline": "1098",
        "healthcare": "104",
        "tollfree": "1800-000-000"
    }

    # 🎨 Stylish Header
    def header():
        print("\n" + "═" * 45)
        print("📒✨  SMART CONTACT BOOK  ✨📒".center(45))
        print("═" * 45)

    # 📋 Menu
    def menu():
        print("\n🎯 Choose an option:")
        print("  1️⃣  ➤ Add Contact")
        print("  2️⃣  ➤ Search Contact")
        print("  3️⃣  ➤ Show All Contacts")
        print("  4️⃣  ➤ Exit")

    # ➕ Add Contact
    def add_contact():
        print("\n➕ ADD NEW CONTACT")
        print("─" * 30)
        name = input("👤 Enter name : ").strip().lower()
        phone = input("📱 Enter phone: ").strip()

        contacts[name] = phone
        search_contacts[name] = phone

        print("\n✅ Contact saved successfully!\n")

    # 🔍 Smart Search (partial match)
    def search_contact():
        print("\n🔍 SEARCH CONTACT")
        print("─" * 30)

        print("\n🚨 Emergency Helplines:")
        print("  police | fire | ambulance")
        print("  women helpline | child helpline")
        print("  healthcare | tollfree\n")

        query = input("🔎 Enter name: ").strip().lower()

        found = False

        print("\n📇 RESULTS")
        print("═" * 30)

        for name, phone in search_contacts.items():
            if query in name:   # 🔥 partial match algorithm
                print(f"👤 {name.title():<20} 📞 {phone}")
                found = True

        if not found:
            print("❌ No matching contact found")

        print("")

    # 📜 Show All Contacts
    def show_all():
        print("\n📜 ALL CONTACTS")
        print("═" * 35)

        if not search_contacts:
            print("⚠️ No contacts available\n")
            return

        for name, phone in sorted(search_contacts.items()):
            print(f"👤 {name.title():<20} 📞 {phone}")

        print("")

    # 🔁 Main Loop
    while True:
        header()
        menu()

        choice = input("\n👉 Enter choice: ").strip()

        if choice == "1":
            add_contact()
        elif choice == "2":
            search_contact()
        elif choice == "3":
            show_all()
        elif choice == "4":
            print("\n👋 Exiting... Stay safe! 💙\n")
            break
        else:
            print("\n⚠️ Invalid choice! Try again.\n")


# ▶️ Run the app
contact_book()









```

#                                            🚀 FULL SECURE EPIC CONTACT BOOK 🚀

```python
from colorama import Fore, Style, init

init(autoreset=True, convert=True, strip=False)

# 🔐 Default password
PASSWORD = "1234"

def contact_book():
    contacts = {}

    # 🚨 Emergency Contacts
    search_contacts = {
        "police": "100",
        "fire": "101",
        "ambulance": "102",
        "women helpline": "1091",
        "child helpline": "1098",
        "healthcare": "104",
        "tollfree": "1800-000-000"
    }

    # 🎨 UI Components
    def header():
        print(Fore.CYAN + "\n" + "═" * 50)
        print(Fore.MAGENTA + "🔐 SMART SECURE CONTACT BOOK 🔐".center(50))
        print(Fore.CYAN + "═" * 50)

    def menu():
        print(Fore.YELLOW + "\n🎯 DASHBOARD")
        print(Fore.GREEN + "  1️⃣  Add Contact")
        print(Fore.GREEN + "  2️⃣  Search Contact")
        print(Fore.GREEN + "  3️⃣  Show All Contacts")
        print(Fore.RED   + "  4️⃣  Exit")
        print(Fore.GREEN + "  5️⃣  Settings")


    # 🔐 Login System
    def login():
        print(Fore.BLUE + "\n🔐 LOGIN REQUIRED")
        for attempt in range(3):
            pwd = input("Enter Password or 4 for Quit: ")

            if pwd == PASSWORD:
                print(Fore.GREEN + "\n✅ Access Granted!\n")
                return True
            elif pwd == "4":
                print("👋 Come Back Soon")
                return False
            else:
                print(Fore.RED + "❌ Wrong Password! Try Again")
                print("💡 Hint: A simple number sequence")

        print(Fore.RED + "\n🚫 Too many failed attempts.")
        return False

    # ➕ Add Contact
    def add_contact():
        print(Fore.CYAN + "\n➕ ADD CONTACT")
        print("─" * 30)

        name = input("👤 Name : ").strip().lower()

        while True:
            phone = input("📱 Enter Phone Number (10 digits): ").strip()

            if phone.isdigit() and len(phone) == 10:
                contacts[name] = phone
                search_contacts[name] = phone
                print(Fore.GREEN + "\n✅ Contact Saved to Contact Book\n")
                break
            else:
                print(Fore.RED + "❌ Invalid phone number! Must be exactly 10 digits.")

    # 🔍 Search Contact
    def search_contact():
        print(Fore.CYAN + "\n🔍 SEARCH CONTACT")
        print("─" * 30)

        print(Fore.YELLOW + "\n🚨 Emergency:")
        print(" police | fire | ambulance | women helpline")
        print(" child helpline | healthcare | tollfree\n")

        query = input("🔎 Search: ").strip().lower()

        print(Fore.MAGENTA + "\n📇 RESULTS")
        print("═" * 30)

        found = False
        for name, phone in search_contacts.items():
            if query in name:
                print(Fore.WHITE + f"👤 {name.title():<20} 📞 {phone}")
                found = True

        if not found:
            print(Fore.RED + "❌ No contact found")

        print()

    # 📜 Show All Contacts
    def show_all():
        print(Fore.CYAN + "\n📜 ALL CONTACTS")
        print("═" * 35)

        for name, phone in sorted(search_contacts.items()):
            print(Fore.WHITE + f"👤 {name.title():<20} 📞 {phone}")

        print()

    # ⚙️ Settings
    def settings_menu():
        while True:
            print(Fore.YELLOW + "\n⚙️ SETTINGS")
            print("👉 1 Block Number")
            print("👉 2 Contact Ringtone")
            print("👉 3 Back")

            choice = input("Enter option: ")

            if choice == "1":
                number = input("Enter number to block: ")
                print(Fore.GREEN + f"{number} is blocked ✅")

            elif choice == "2":
                number = input("Enter Contact Number: ")

                print("🎵 1 Custom Ringtone")
                print("🎵 2 Horizon Ringtone")
                print("🎵 3 Xylophone Ringtone")

                ringtone = input("Choose ringtone: ")

                if ringtone == "1":
                    print(Fore.GREEN + f"Custom ringtone set for {number} ✅")
                elif ringtone == "2":
                    print(Fore.GREEN + f"Horizon ringtone set for {number} ✅")
                elif ringtone == "3":
                    print(Fore.GREEN + f"Xylophone ringtone set for {number} ✅")
                else:
                    print(Fore.RED + "Invalid Choice")

            elif choice == "3":
                break
            else:
                print(Fore.RED + "Invalid Choice")

    if not login():
        return

    while True:
        header()
        menu()

        print(Fore.YELLOW + "\n👉 Enter choice: ", end="")
        choice = input()

        if choice == "1":
            add_contact()
        elif choice == "2":
            search_contact()
        elif choice == "3":
            show_all()
        elif choice == "4":
            print(Fore.CYAN + "\n👋 Goodbye! Stay safe 💙\n")
            break
        elif choice == "5":
            settings_menu()
        else:
            print(Fore.RED + "\n⚠️ Invalid choice\n")


# ▶️ Run App
contact_book()
```

#                                            🚀 1. 🔐 PASSWORD MANAGER (Very Useful 🔥)

```python
from colorama import Fore, init
init(autoreset=True)

MASTER_KEY = "admin"

def password_manager():
    vault = {}

    def login():
        print(Fore.CYAN + "\n🔐 PASSWORD VAULT LOGIN")
        for _ in range(3):
            if input("Enter Master Key: ") == MASTER_KEY:
                print(Fore.GREEN + "✅ Access Granted\n")
                return True
            print(Fore.RED + "❌ Wrong Key")
        return False

    def add():
        site = input("🌐 Website: ").lower()
        pwd = input("🔑 Password: ")
        vault[site] = pwd
        print(Fore.GREEN + "✅ Saved!\n")

    def view():
        print("\n📂 STORED PASSWORDS")
        print("═" * 30)
        for k, v in vault.items():
            print(f"{k:<15} 🔒 {'*' * len(v)}")
        print()

    def search():
        key = input("🔍 Search: ").lower()
        print("\nRESULT:")
        for k, v in vault.items():
            if key in k:
                print(f"{k} → {v}")

    if not login():
        return

    while True:
        print("\n1.Add  2.View  3.Search  4.Exit")
        c = input("👉 ")
        if c == "1": add()
        elif c == "2": view()
        elif c == "3": search()
        elif c == "4": break

password_manager()
```

#                                        🚀 2. 💰 SMART EXPENSE TRACKER (Super Practical 💸)

```python
from colorama import Fore, init
init(autoreset=True)

def expense_tracker():
    expenses = []

    def add():
        name = input("📝 Item: ")
        amt = float(input("💰 Amount: "))
        expenses.append((name, amt))
        print(Fore.GREEN + "✅ Added!")

    def summary():
        total = sum(x[1] for x in expenses)
        print(Fore.CYAN + f"\n💵 Total Spent: ₹{total}")

        print("\n📊 Breakdown:")
        for item, amt in expenses:
            print(f"{item:<15} ₹{amt}")

    while True:
        print("\n1.Add Expense\n2.View Summary\n3.Exit")
        c = input("👉 ")

        if c == "1": add()
        elif c == "2": summary()
        elif c == "3": break

expense_tracker()
```

#                           3 🗳️ Voting System (Real-life: Poll system)


```python
votes = {
    "Alice": 0,
    "Bob": 0,
    "Charlie": 0
}

while True:
    vote = input("Vote (or 'end'): ")

    if vote == "end":
        break

    if vote in votes:
        votes[vote] += 1
    else:
        print("❌ Invalid candidate")

print("\nResults:", votes)

winner = max(votes, key=votes.get)
print("🏆 Winner is:", winner)
```

#                           🚀  🧠 SMART TO-DO LIST (Daily Life Game-Changer ✅)

```python
from colorama import Fore, init
init(autoreset=True)

def todo():
    tasks = []

    def add():
        t = input("📝 Task: ")
        tasks.append([t, False])

    def view():
        print("\n📋 TASK LIST")
        print("═" * 30)
        for i, (t, done) in enumerate(tasks):
            status = "✅" if done else "❌"
            print(f"{i+1}. {t} {status}")

    def complete():
        view()
        i = int(input("✔ Task number: ")) - 1
        tasks[i][1] = True

    while True:
        print("\n1.Add\n2.View\n3.Complete\n4.Exit")
        c = input("👉 ")

        if c == "1": add()
        elif c == "2": view()
        elif c == "3": complete()
        elif c == "4": break

todo()
```

# 🚀 1. 🏦 MINI BANK SYSTEM (ATM Simulation)

```python
from colorama import Fore, init
init(autoreset=True)

PIN = "1234"

def bank():
    balance = 1000

    def login():
        print(Fore.CYAN + "\n🏦 ATM LOGIN")
        for _ in range(3):
            if input("Enter PIN: ") == PIN:
                print(Fore.GREEN + "✅ Access Granted\n")
                return True
            print(Fore.RED + "❌ Wrong PIN")
        return False

    def deposit():
        nonlocal balance
        amt = float(input("💰 Enter amount: "))
        balance += amt
        print(Fore.GREEN + f"✅ Deposited! Balance: ₹{balance}")

    def withdraw():
        nonlocal balance
        amt = float(input("💸 Enter amount: "))
        if amt <= balance:
            balance -= amt
            print(Fore.GREEN + f"✅ Withdrawn! Balance: ₹{balance}")
        else:
            print(Fore.RED + "❌ Insufficient balance")

    def check():
        print(Fore.YELLOW + f"💵 Balance: ₹{balance}")

    if not login():
        return

    while True:
        print("\n1.Deposit 2.Withdraw 3.Check Balance 4.Exit")
        c = input("👉 ")

        if c == "1": deposit()
        elif c == "2": withdraw()
        elif c == "3": check()
        elif c == "4": break

bank()
```

# 🚀 🎓 STUDENT MANAGEMENT SYSTEM

```python
from colorama import Fore, init
init(autoreset=True)

def student_system():
    students = {}

    def add():
        name = input("👤 Name: ")
        marks = list(map(int, input("📊 Marks (space separated): ").split()))
        students[name] = marks
        print(Fore.GREEN + "✅ Added!")

    def report():
        print("\n📋 REPORT CARD")
        print("═" * 35)
        for name, marks in students.items():
            avg = sum(marks)/len(marks)
            grade = "A" if avg >= 80 else "B" if avg >= 60 else "C"
            print(f"{name:<10} Avg:{avg:.1f} Grade:{grade}")

    while True:
        print("\n1.Add Student\n2.View Report\n3.Exit")
        c = input("👉 ")

        if c == "1": add()
        elif c == "2": report()
        elif c == "3": break

student_system()
```

# 🚀 Fixed + Upgraded “Team Battle Voting Game”


```python
votes = {"A": 0, "B": 0, "C": 0}
target = 20
chances_per_user = 5

print("⚔️  WELCOME TO THE VOTING BATTLE ARENA ⚔️")
print("🏳️ Choose your : partner:---> Elon musk for E / Mukesh Ambhani for M/ Mark Zuckherbherg for MZ")
print("🎯 First team to reach target wins!\n")

# 🧑 User selects team
team = input("🔥 Enter your team: ").upper().strip()

if team not in votes:
    print("❌ Invalid team!")
else:
    print(f"\n🎖️ You joined Team {team}! Let the battle begin...\n")

    # 🔁 User chances
    for chance in range(1, chances_per_user + 1):
        print(f"\n🎯 Chance {chance}/{chances_per_user}")

        vote = input("🗳️ Vote your team (or type 'end'): ").upper().strip()

        if vote == "END":
            break

        if vote in votes:
            votes[vote] += 1
            print("✅ Vote Counted!")
        else:
            print("❌ Invalid vote!")

        # 🤖 Opponent auto votes
        for opponent in votes:
            if opponent != team:
                votes[opponent] += 1

        # 📊 Stylish Scoreboard
        print("\n📊 BATTLE SCOREBOARD 📊")
        print("=================================")

        for t in votes:
            print(f"🏳️ Team {t} | ", end="")
            
            # 🔹 nested loop for bar display
            for i in range(votes[t]):
                print("█", end=" ")
            
            print(f"({votes[t]})")

        print("=================================")

        # 🏆 Check winner
        for t in votes:
            if votes[t] >= target:
                print(f"\n🏆 TEAM {t} HAS CONQUERED THE ARENA! 🎉")
                exit()

print("\n⚡ Battle Ended!")
```

# 🚀 Final “Elite Voting Battle Arena”


```python
import time

# 🧠 Team Mapping
teams = {
    "E": "Elon Musk",
    "M": "Mukesh Ambani",
    "MZ": "Mark Zuckerberg"
}

votes = {key: 0 for key in teams}

target = 15
chances_per_user = 5

# 🎬 Start Time
start_time = time.time()

def show_intro():
    print("⚔️  WELCOME TO THE VOTING BATTLE ARENA ⚔️")
    print("🏳️ Choose your partner:")
    print("   E  ➤ Elon Musk 🚀")
    print("   M  ➤ Mukesh Ambani 💼")
    print("   MZ ➤ Mark Zuckerberg 🌐")
    print(f"\n🎯 Target Votes: {target}")
    print("🔥 Let the battle begin!\n")

def display_scoreboard():
    print("\n📊 LIVE SCOREBOARD 📊")
    print("=================================")

    for key in votes:
        print(f"🏳️ {teams[key]:<18} | ", end="")

        # 🔹 Nested loop for bar style
        for i in range(votes[key]):
            print("█", end=" ")

        print(f"({votes[key]})")

    print("=================================")

def show_time():
    end_time = time.time()
    duration = round(end_time - start_time, 2)
    print(f"\n⏱️ Total Voting Time: {duration} seconds")

# 🎮 Game Start
show_intro()

user_team = input("🔥 Enter your team (E/M/MZ): ").upper().strip()

if user_team not in teams:
    print("❌ Invalid team! GAME OVER 💀")
else:
    print(f"\n🎖️ You joined {teams[user_team]}!\n")

    for chance in range(1, chances_per_user + 1):
        print(f"🎯 Chance {chance}/{chances_per_user}")

        vote = input("🗳️ Cast your vote (or 'END'): ").upper().strip()

        if vote == "END":
            print("\n💀 GAME OVER — You exited the arena!")
            break

        if vote in votes:
            votes[vote] += 1
            print("✅ Vote Counted!")
        else:
            print("❌ Invalid vote!")

        # 🤖 Opponent auto votes
        for key in votes:
            if key != user_team:
                votes[key] += 1

        display_scoreboard()

        # 🏆 Check winner
        for key in votes:
            if votes[key] >= target:
                print(f"\n🏆 {teams[key]} DOMINATES THE ARENA! 🔥")
                show_time()
                exit()

    print("\n💀 GAME OVER — Chances exhausted!")

# ⏱️ Show total time
show_time()
```

```python
import time
import sys

# 🧠 Team Mapping
teams = {
    "E": "Elon Musk",
    "M": "Mukesh Ambani",
    "MZ": "Mark Zuckerberg"
}

votes = {key: 0 for key in teams}

target = 15
chances_per_user = 5

# 🎬 Start Time
start_time = time.time()

def show_intro():
    print("⚔️  WELCOME TO THE VOTING BATTLE ARENA ⚔️")
    print("🏳️ Choose your partner:")
    print("   E  ➤ Elon Musk 🚀")
    print("   M  ➤ Mukesh Ambani 💼")
    print("   MZ ➤ Mark Zuckerberg 🌐")
    print(f"\n🎯 Target Votes: {target}")
    print("🔥 Let the battle begin!\n")

def display_scoreboard():
    print("\n📊 LIVE SCOREBOARD 📊")
    print("=================================")

    for key in votes:
        print(f"🏳️ {teams[key]:<18} | ", end="")

        # 🔹 Nested loop for bar style
        for i in range(votes[key]):
            print("█", end=" ")

        print(f"({votes[key]})")

    print("=================================")

def show_time():
    end_time = time.time()
    duration = round(end_time - start_time, 2)
    print(f"\n⏱️ Total Voting Time: {duration} seconds")

# 🎮 Game Start
show_intro()

user_team = input("🔥 Enter your team (E/M/MZ): ").upper().strip()

if user_team not in teams:
    print("❌ Invalid team! GAME OVER 💀")
else:
    print(f"\n🎖️ You joined {teams[user_team]}!\n")

    for chance in range(1, chances_per_user + 1):
        print(f"🎯 Chance {chance}/{chances_per_user}")

        vote = input("🗳️ Cast your vote (or 'END'): ").upper().strip()

        if vote == "END":
            print("\n💀 GAME OVER — You exited the arena!")
            break

        if vote in votes:
            votes[vote] += 1
            print("✅ Vote Counted!")
        else:
            print("❌ Invalid vote!")

        # 🤖 Opponent auto votes
        for key in votes:
            if key != user_team:
                votes[key] += 1

        display_scoreboard()

        # 🏆 Check winner
        for key in votes:
            if votes[key] >= target:
                print(f"\n🏆 {teams[key]} DOMINATES THE ARENA! 🔥")
                show_time()
                sys.exit()   # ✅ Correct termination

    print("\n💀 GAME OVER — Chances exhausted!")

# ⏱️ Show total time
show_time()
```

```python
counter = 0

while True:
    num = input("Enter a number (or 'exit' to stop): ")

    if num.lower() == "exit":
        print("Exiting program.")
        break

    try:
        num = int(num)
    except ValueError:
        print("Please enter a valid number!")
        continue

    # Condition 1: number is greater than 10
    if num > 10:
        print("Condition 1 met: number > 10")
        counter += 1

    # Condition 2: number is even
    if num % 2 == 0:
        print("Condition 2 met: number is even")
        counter += 1

    # If both conditions satisfied (counter >= 2), raise an error
    if counter >= 2:
        raise Exception("🚨 Both conditions satisfied! Program stops with an error.")
```

```python
total = 0

while True:
    num = input("Enter a number to add: ")

    if num.lower() == "exit":
        print("Final total:", total)
        break

    try:
        num = int(num)
    except ValueError:
        print("Invalid number!")
        continue

    total += num
    print("Current total:", total)

    # Condition 1: total > 50
    # Condition 2: last number entered is even
    if total > 50 and num % 2 == 0:
        raise ValueError("🚨 Total exceeded 50 with even number! Error triggered.")
```
