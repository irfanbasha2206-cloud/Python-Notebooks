# Standard Library Modules Table

| Category                         | Examples                                  | Purpose                                                    |
| -------------------------------- | ----------------------------------------- | ---------------------------------------------------------- |
| **File and OS operations**       | `os`, `shutil`, `pathlib`, `glob`         | File handling, directory operations, path manipulation     |
| **Data handling & formats**      | `json`, `csv`, `configparser`, `pickle`   | Reading/writing JSON, CSV, INI files, object serialization |
| **Networking & Internet**        | `urllib`, `http`, `socket`, `ftplib`      | Web requests, HTTP servers, FTP, sockets                   |
| **Data compression / archiving** | `zipfile`, `tarfile`, `gzip`, `bz2`       | Zip, tar, and other compressed files                       |
| **Date and time**                | `datetime`, `time`, `calendar`            | Date, time, calendar operations                            |

| **Text processing**              | `re`, `string`, `textwrap`                | Regular expressions, string utilities, formatting          |

```python
# OS Module
'''

The os module in Python is a built-in library that allows your program to interact with the operating system (OS). 
It provides a way to work with files, directories, environment variables, and system-level operations in a platform-independent way.

'''
# The os module acts as a bridge between Python and the underlying OS (like Windows, Linux, macOS). It helps you:
'''
 1. Work with Files & Directories
Create, delete, rename files/folders
Navigate directory structures

 2. Handle File Paths
Join paths safely
Get absolute/relative paths

 3. Access Environment Variables
Read system variables like PATH, HOME

 4. Execute System Commands
Run shell commands from Python

 5. Get System Information
OS name, current working directory, etc.

'''

# Key Concepts & Functions
'''

 1. Directory Operations
os.getcwd() → Get current working directory
os.chdir(path) → Change directory
os.listdir(path) → List files in a directory
os.mkdir("folder") → Create a directory
os.rmdir("folder") → Remove a directory
os.makedirs(name, mode=0o777, exist_ok=False) → create directories recursively

name → Path of the directory to create
mode → Permission (default: 0o777)
exist_ok →
          False → Raises error if directory already exists
          True → Ignores error if directory exists


 2. File Operations
os.remove("file.txt") → Delete a file
os.rename("old.txt", "new.txt") → Rename file


 3. Path Handling (os.path)
os.path.exists(path) → Check if path exists
os.path.join(a, b) → Join paths safely
os.path.abspath(path) → Get absolute path
os.path.isfile() → skip directories.



 4. Environment Variables
os.environ → Dictionary of environment variables
os.getenv("VAR_NAME") → Get variable


 5. System Info
os.name → OS type ('nt' for Windows, 'posix' for Linux/macOS)
```

```python
# Automate the errors


path = r"C:\Users\ELCOT\PYTHON\PythonOS"
files_list = os.listdir(path)
for i in files_list:
    file_path = os.path.join(path,i)
    file = open(file_path,"r")
    text = file.read()
    file.close()
    correct_text = text.replace("PY", "Python")
    file = open(file_path,"w")
    file.write(correct_text)
    file.close()
    print(correct_text)

    
```
