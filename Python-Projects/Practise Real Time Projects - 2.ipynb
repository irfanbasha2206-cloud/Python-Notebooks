# AI “JARVIS” Voice Assistant 

```python
import pyttsx3
import datetime
import webbrowser

# Initialize text-to-speech engine
engine = pyttsx3.init()

# Speak function
def speak(text):
    print("Jarvis:", text)
    try:
        engine.say(text)
        engine.runAndWait()
    except:
        # Prevent crash if audio doesn't work in Jupyter
        pass

# Take user input
def listen():
    return input("You: ").lower()

# Main assistant function
def run_jarvis():
    speak("Hello boss, I am Jarvis.")

    while True:
        command = listen()

        if "time" in command:
            current_time = datetime.datetime.now().strftime("%H:%M")
            speak(f"The time is {current_time}")

        elif "date" in command:
            today = datetime.datetime.now().strftime("%Y-%m-%d")
            speak(f"Today's date is {today}")

        elif "open youtube" in command:
            webbrowser.open("https://www.youtube.com")
            speak("Opening YouTube")

        elif "open google" in command:
            webbrowser.open("https://www.google.com")
            speak("Opening Google")

        elif "exit" in command or "quit" in command:
            speak("Goodbye boss")
            break

        elif command.strip() == "":
            speak("I didn't catch that")

        else:
            speak("Sorry, I don't understand that yet")

# Run the assistant
run_jarvis()
```

# AI Mood-Based Music Recommender

```python
import builtins
import time

# Main moods with songs
moods = {
    "Feel Good": ["Shape of You", "Happy - Pharrell", "On Top of the World"],
    "Emotional": ["Let Her Go", "Someone Like You", "Fix You"],
    "Motivated": ["Believer", "Lose Yourself", "Stronger"],
}

# Secondary moods with songs by language
mood_2 = {
    "Romantic": [
        {"Hindi": ["Tum Hi Ho", "Raataan"]},
        {"English": ["Perfect", "All of Me"]},
        {"Tamil": ["Munbe Vaa", "Vaseegara"]},
        {"Malayalam": ["Malare", "Darshana"]}
    ],
    "Friendship": [
        {"English": ["Count on Me", "See You Again"]},
        {"Tamil": ["Mustafa Mustafa", "En Friend Pola"]},
        {"Hindi": ["Yeh Dosti", "Dil Chahta Hai"]},
        {"Malayalam": ["Aluva Puzha", "Onakka Munthiri"]}
    ],
    "Energetic": [
        {"Tamil": ["Vaathi Coming", "Surviva"]},
        {"English": ["Uptown Funk", "Wake Me Up"]},
        {"Malayalam": ["Butta Bomma", "Jimikki Kammal"]},
        {"Hindi": ["Dhoom Machale", "Kala Chashma"]}
    ]
}

def normalize_input(user_input):
    """Remove extra spaces and normalize capitalization"""
    return ' '.join(user_input.strip().split()).title()

def show_moods():
    print("\nAvailable Primary Moods:")
    for mood in moods:
        print("-", mood)
        time.sleep(1.0)
    print("- Exit")
    time.sleep(0.5)

def show_secondary_moods():
    print("\nAvailable Secondary Moods:")
    for mood in mood_2:
        print("-", mood)
        time.sleep(0.5)
    print("- Back")
    time.sleep(0.5)
    print("- Exit")
    time.sleep(0.5)

def recommend():
    while True:
        show_moods()
        mood_input = builtins.input("\nEnter your mood (or type 'Exit'): ")
        mood = normalize_input(mood_input)

        if mood.lower() == "exit":
            print("\n👋 Exiting the music program. Goodbye!")
            time.sleep(0.5)
            break

        if mood in moods:
            print(f"\n🎧 Songs in {mood} mood:")
            time.sleep(0.5)
            for song in moods[mood]:
                print("🎵", song)
                time.sleep(0.5)

            while True:
                print("\nSwitch to secondary moods or go back:")
                show_secondary_moods()
                sec_input = builtins.input("\nEnter secondary mood: ")
                sec_mood = normalize_input(sec_input)

                if sec_mood.lower() == "exit":
                    print("\n👋 Exiting the music program. Goodbye!")
                    time.sleep(0.5)
                    return
                elif sec_mood.lower() == "back":
                    break

                if sec_mood in mood_2:
                    print(f"\n🎶 Songs in {sec_mood} mood by language:")
                    time.sleep(0.5)
                    for lang_dict in mood_2[sec_mood]:
                        for lang, songs in lang_dict.items():
                            print(f"\n{lang} Songs:")
                            time.sleep(0.5)
                            for song in songs:
                                print("🎵", song)
                                time.sleep(0.5)
                else:
                    print("\n❌ Invalid secondary mood! Try again.")
                    time.sleep(0.5)
        else:
            print("\n❌ Invalid primary mood! Please choose from the list.")
            time.sleep(0.5)

# Run the interactive program
recommend()
```

# 🎵 Trendy Music Media: 🎶 "BeatPulse" 🎶


```python
import builtins
import pygame
import time
import os
from colorama import Fore, Style, init

# Initialize colorama
init(autoreset=True)

# Initialize Pygame mixer
pygame.mixer.init()

# Music library categorized by moods
moods = {
    "Feel Good": [
        "songs/shape_of_you.mp3",
        "songs/happy_pharrell.mp3",
        "songs/on_top_of_the_world.mp3"
    ],
    "Emotional": [
        "songs/let_her_go.mp3",
        "songs/someone_like_you.mp3",
        "songs/fix_you.mp3"
    ],
    "Motivated": [
        "songs/believer.mp3",
        "songs/lose_yourself.mp3",
        "songs/stronger.mp3"
    ],
    "Romantic": [
        "songs/tum_hi_ho.mp3",
        "songs/perfect.mp3",
        "songs/munbe_vaa.mp3",
        "songs/malare.mp3"
    ],
    "Friendship": [
        "songs/count_on_me.mp3",
        "songs/yeh_dosti.mp3",
        "songs/aluva_puzha.mp3"
    ],
    "Energetic": [
        "songs/vaathi_coming.mp3",
        "songs/uptown_funk.mp3",
        "songs/dhoom_machale.mp3"
    ]
}

# Function to display moods with styling
def show_moods():
    print(Fore.CYAN + Style.BRIGHT + "\n🎧 Available Moods in BeatPulse 🎧")
    for mood in moods:
        print(Fore.YELLOW + " ➤ " + mood)

# Function to display commands
def show_commands():
    print(Fore.MAGENTA + "\n🎵 Commands: play | pause | resume | next | loop | stop | exit")

# Function to play music with controls
def play_music(song_list):
    current = 0
    loop = False
    while True:
        song_path = song_list[current]
        if not os.path.exists(song_path):
            print(Fore.RED + f"❌ Song file not found: {song_path}")
            current += 1
            if current >= len(song_list): break
            continue

        pygame.mixer.music.load(song_path)
        pygame.mixer.music.play()
        print(Fore.GREEN + Style.BRIGHT + f"\n🎶 Now Playing: {os.path.basename(song_path).replace('_', ' ').replace('.mp3','')} 🎶")

        while pygame.mixer.music.get_busy() or loop:
            command = builtins.input(Fore.CYAN + "\nEnter command: ").lower().strip()
            if command == "pause":
                pygame.mixer.music.pause()
                print(Fore.YELLOW + "⏸ Paused")
            elif command == "resume" or command == "play":
                pygame.mixer.music.unpause()
                print(Fore.GREEN + "▶ Resumed")
            elif command == "next":
                pygame.mixer.music.stop()
                current += 1
                if current >= len(song_list):
                    print(Fore.MAGENTA + "⚡ End of playlist")
                    return
                break
            elif command == "loop":
                loop = not loop
                print(Fore.BLUE + ("🔁 Looping enabled" if loop else "❌ Looping disabled"))
            elif command == "stop":
                pygame.mixer.music.stop()
                print(Fore.RED + "⏹ Stopped playback")
                return
            elif command == "exit":
                pygame.mixer.music.stop()
                print(Fore.RED + "👋 Exiting BeatPulse")
                exit()
            else:
                print(Fore.YELLOW + "❌ Unknown command. Use play/pause/resume/next/loop/stop/exit")

# Main function
def run_beatpulse():
    print(Fore.MAGENTA + Style.BRIGHT + "\n✨ Welcome to BeatPulse - Your Trendy Music Media ✨")
    while True:
        show_moods()
        mood_choice = builtins.input(Fore.CYAN + "\nSelect your mood: ").title().strip()
        if mood_choice in moods:
            print(Fore.GREEN + f"\n✅ Loading {mood_choice} playlist...")
            time.sleep(1)
            play_music(moods[mood_choice])
        else:
            print(Fore.RED + "❌ Invalid mood! Try again.")

# Run the media
run_beatpulse()
```

# 🐐 Aadu Puli Aattam 🐅

```python
import time

board = ["_" for _ in range(9)]
tigers = [0, 8]
goats = []

for t in tigers:
    board[t] = "T"

def print_board():
    print("\nBoard:")
    for i in range(0, 9, 3):
        print(board[i], board[i+1], board[i+2])

def safe_input(msg):
    try:
        return int(input(msg))
    except:
        print("❌ Enter a number!")
        return None

def game():
    print("🐐🐅 Aadu Puli Aattam")

    turn = "goat"
    goat_count = 0

    while True:
        print_board()

        if turn == "goat":
            if goat_count < 6:
                pos = safe_input("Place Goat (0-8): ")
                if pos is None or pos not in range(9):
                    continue

                if board[pos] == "_":
                    board[pos] = "G"
                    goats.append(pos)
                    goat_count += 1
                    turn = "tiger"
                else:
                    print("❌ Spot already taken!")

        else:
            print("Tiger move")
            old = safe_input("Select Tiger position: ")
            new = safe_input("Move to position: ")

            if old is None or new is None:
                continue

            if old in tigers and board[new] == "_":
                board[old] = "_"
                board[new] = "T"
                tigers.remove(old)
                tigers.append(new)
                turn = "goat"
            else:
                print("❌ Invalid move!")

        if goat_count >= 6:
            print("\nGame demo finished ✅")
            break

game()
```

# Digital World Clock

```python
import datetime
import pytz
import time
import os

# List of cities and their timezones
world_timezones = {
    "New York": "America/New_York",
    "London": "Europe/London",
    "Paris": "Europe/Paris",
    "Tokyo": "Asia/Tokyo",
    "Sydney": "Australia/Sydney",
    "Delhi": "Asia/Kolkata",
}

def display_world_clock():
    while True:
        os.system('cls' if os.name == 'nt' else 'clear')  # Clear screen for live clock
        print("🌏 Digital World Clock 🌏\n")
        now_utc = datetime.datetime.now(pytz.utc)
        for city, tz_name in world_timezones.items():
            tz = pytz.timezone(tz_name)
            city_time = now_utc.astimezone(tz)
            print(f"{city}: {city_time.strftime('%Y-%m-%d %H:%M:%S')} ({tz_name})")
        time.sleep(1)

if __name__ == "__main__":
    display_world_clock()
```
