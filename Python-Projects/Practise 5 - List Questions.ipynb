```python
#write a program to find the difference between the sum of all even elements abd the sum of all odd elements from the given array A
A = [56,63,87,24,32,13,15,19,44,52]
even_sum = 0
odd_sum = 0
for i in A:
    if i%2==0:
        even_sum = even_sum + i
    else:
        odd_sum = odd_sum + i
print("Even Sum ",even_sum)
print("Odd Sum:",odd_sum)
print("Difference of Even_sum & Odd_sum:",even_sum-odd_sum)        
```

```python
#ceil and floor by given array and integer
a = [-2,4,4,8,9]    # to find the first smallest and first largest of an given integer 
b = 0 
if b in a:
    print(b,b)    
else:
    a.append(b)
    a.sort()
    for i in range(len(a)):
        if a[i] == b:
            c = [a[i-1],a[i+1]]
            print(a)
            print(c)
```

```python
#ceil and floor by given array and integer
#Alternate method
a = [-2,4,4,8,9]    # to find the first smallest and first largest of an given integer 
b = 0 
if b in a:
    print(b,b)
else:
    a.append(b)
    a.sort()
    index = a.index(b)
    c = [a[index - 1] , a[index +1]]
    print(c)
```

```python
'''inverse of the array
input --> a = [2,0,1] and output ----> [1,2,0]
     index --> 0,1,2
     swaping the index and value and index goes for position of value'''

a = [2,0,1]
b = [0] * len(a)
for i in range(len(a)):
    b[a[i]] = i    # here the itertion value will be used as a index value which is in 1st iteration "a[i]" --> a[0] then b[a[i]] --> b[a[0]] --> b[2] and b[2](index) = i(0)(value) it is equal to every iteration
print(b)
```

```python
a = [1,2,3,4,5,6]      # if list contains in even order
sub = - 1
for i in range(len(a)//2):
    a[i] , a[sub] = a[sub] , a[i]
    sub = sub - 1
print(a)
```

```python
a = [1, 2, 3, 4, 5]   # if list contains in odd order
n = len(a)
for i in range(n // 2):
    a[i], a[n - i - 1] = a[n - i - 1], a[i]

print(a)
```
