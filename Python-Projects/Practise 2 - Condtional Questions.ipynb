```python
##### student grade program
tamil=int(input("Enter the Tamil Mark"))
english=int(input("Enter the English Mark"))
maths=int(input("Enter the Maths Mark"))
science=int(input("Enter the Science Mark"))
social=int(input("Enter the Social Mark"))
total=tamil+english+maths+science+social
average=total//5
print("The total is",total,"out of 500")
print("The average is",average,"Out of 100")
if (average>=90) and (average<=100):
    print("A Grade")
    print("Very Good Student")
elif (average>=80) and (average<90):
    print("B Grade")
    print("Good Student")
elif (average>=70) and (average<80):
    print("C Grade")
    print("Average Student")
else:
    print("D Grade")
    print("Poor Student")
    
```

```python
#library management ------> My code
days=int(input("Enter how many days: "))
if days>=1 and days<=10:
    perday_fine=0.5*days
    print("Fine is",int(perday_fine),"rs")
elif days>=11 and days<=20:
    perday_fine=1*days
    print("Fine is",perday_fine,"rs")
elif days>=21 and days<=31:
    perday_fine=5*days
    print("Fine is",perday_fine,"rs")
else:
    print("Membership cancelled")
```

```python
# Library Fine Calculation Program ----> Chatgpt code

book_name = input("Enter Book Name: ")
days = int(input("Enter number of overdue days: "))

if days <= 0:
    fine = 0

else:
    if days >= 1 and days <= 10:
        fine = days * 0.5

    else:
        if days >= 11 and days <= 20:
            fine = days * 1

        else:
            if days >= 21 and days <= 31:
                fine = days * 5
            else:
                fine = 0
                print("Fine applicable only up to 31 days.")

print("Book Name:", book_name)
print("Overdue Days:", days)
print("Total Fine: ₹", fine)


```
